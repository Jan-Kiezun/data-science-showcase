```python
import matplotlib.pyplot as plt
import random
from aco import AntColony
```

```python
plt.style.use("dark_background")
COORDS = (
    (20, 52),
    (43, 50),
    (20, 84),
    (70, 65),
    (29, 90),
    (87, 83),
    (73, 23),
    (55, 50),
    (55, 20),
    (44, 70),
    (50, 30),
)
def random_coord():
    r = random.randint(0, len(COORDS))
    return r
```

```python
def plot_nodes(w=12, h=8):
    for x, y in COORDS:
        plt.plot(x, y, "g.", markersize=15)
    plt.axis("off")
    fig = plt.gcf()
    fig.set_size_inches([w, h])


def plot_all_edges():
    paths = ((a, b) for a in COORDS for b in COORDS)

    for a, b in paths:
        plt.plot((a[0], b[0]), (a[1], b[1]))


plot_nodes()
```

```python
from time import perf_counter
def timer(f):
    def wrapper(*args, **kwargs):
        start = perf_counter()
        result = f(*args, **kwargs)
        end = perf_counter()
        print("Time: ", end - start)
        return result
    return wrapper
```

```python
@timer
def train_and_plot():
    colony = AntColony(
        COORDS,
        ant_count=3,
        alpha=0.5,
        beta=1.2,
        pheromone_evaporation_rate=0.1,
        pheromone_constant=1000.0,
        iterations=1000,
    )
    # at=300, al=0.5, be=1.2, ph=0.40, pc=1000.0, it=300, time=37.939235199999985 // wynik nienajlepszy
    # at=600, al=0.5, be=1.2, ph=0.40, pc=1000.0, it=150, time=36.49040379999997 // wyglda lepiej
    # at=600, al=3, be=3, ph=0.20, pc=1000.0, it=150, time=41.27610670000013 // maa rnica pomimo zmiany al i be
    # at=600, al=0.5, be=1.2, ph=0.8, pc=1000.0, it=150, time=40.927097099999855 // maa rnica pomimo zmiany ph
    # at=600, al=0.5, be=1.2, ph=0.8, pc=1000.0, it=30, time=~9.000000000000002 // maa rnica pomimo zmiany it
    # at=600, al=0.5, be=1.2, ph=0.8, pc=1000.0, it=10, time=~3.000000000000001 // maa rnica pomimo zmiany it... znowu
    # at=600, al=0.5, be=1.2, ph=0.8, pc=1000.0, it=3, time=~1.0000000000000002 // ... po co tyle iteracji byo dawa
    # at=600, al=0.5, be=1.2, ph=0.4, pc=1000.0, it=3, time=1.0000000000000002 // ... dla oryginalnego ph wynik podobny
    # at=600, al=0.5, be=1.2, ph=0.4, pc=1000.0, it=1, time=0.4000000000000002 // zawsze ten sam wynik, dobry wynik (dla 1 iteracji)
    # at=600, al=0.5, be=1.2, ph=0.4, pc=1000.0, it=0, time=0.0008831999998619 // Daje None dla it=0
    # at=1, al=0.5, be=1.2, ph=0.1, pc=1000.0, it=1000, time=20.25681050000003 // Daje dobry wynik o dziwo
    # at=3, al=0.5, be=1.2, ph=0.1, pc=1000.0, it=1000, time=20.25681050000003 // Daje super wynik lol



    optimal_nodes = colony.get_path()

    for i in range(len(optimal_nodes) - 1):
        plt.plot(
            (optimal_nodes[i][0], optimal_nodes[i + 1][0]),
            (optimal_nodes[i][1], optimal_nodes[i + 1][1]),
        )

train_and_plot()
plt.show()
```
